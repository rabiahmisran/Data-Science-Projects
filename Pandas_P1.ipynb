```python
# Pandas Data Science Library
import numpy as np
import pandas as pd
print(pd.__version__)
```

```python
# numpy --> array()
arr = np.array([10,20,30,40,50])
print(type(arr))
print(arr)
```

```python
# Pandas Series
# Series is 1D Labelled Array

s = pd.Series([10,20,30,40,50])
print(type(s))
print(s)
print(s.shape)

# In Numpy --> Only Numbers
# In Series --> Numbers + Labels (index)
```

```python
info = pd.Series(['John',28,6700,'Klang'],index=['name','Age','sal','city'])
print(info)
print(info['name'])   # indexing
print(info['Age'])    # getting data values using get()
print(info.iloc[0])   # iloc --> index locatiom

# Series ia like a Dictionary + array combo
```

```python
# values property
# print only values

print(info.values)
```

```python
import pandas as pd

# Indexing Slicing and Using for Loop
s = pd.Series([100,200,300,400,500,600],index=['v1','v2','v3','v4','v5','v6'])
print(s)

# 100  200  300  400  500  600
#  0    1    2    3    4    5
# -6   -5   -4   -3   -2   -1
# v1   v2   v3   v4   v5   v6

# Indexing in Series can be done by using [] or iloc[] or loc[]
print('\nIndexing in Series')
print(s.iloc[0])    # for index numbers
print(s.loc['v5'])  # loc used for Labels
print(s['v2'])      # [] used for Labels
print(s.iloc[-1])

print('\nSlicing in Series')
print(s[1:4])          # 4 will be excluded --> stopIndex
print(s['v1' : 'v5']) # if performing slicing using Labels, the end Label is included

# 100  400  600
print('\nFor Specific Information')
print(s[['v1','v4','v6']])  # for specific information s[['','','']]
print(s[['v1','v2','v3']].values)

# conditional selection
print('\nconditional selection')
print(s[s%3==0])
```

```python
# Using for Loop
for i in s:
  print(i)
```

```python
for k,v in s.items():
  print(k,v)
```

```python
for i in enumerate(s):
  print(i)
```
